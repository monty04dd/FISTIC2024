```python
import pandas as pd
import numpy as np

# Creazione del dataset con errori
np.random.seed(42)  # Per coerenza dei risultati

data = {
    # Variabili numeriche con errori
    "Num1": [np.nan, 3.5, 8.1, -1.2, 10.5, 5.8, "errore", 7.0, 9.2, 6.5, 4.3, 7.7, 8.8, -3.3, "###"],
    "Num2": [2.5, 3.8, 4.1, "?", 5.5, np.nan, 3.3, 4.2, -1.0, 4.9, "N/A", 3.5, 6.1, 5.7, 5.9],
    "Num3": [1.2, -2.5, 3.7, 3.3, "missing", 5.0, 4.1, np.nan, 7.2, 6.3, "!", 3.9, 2.2, 1.1, -0.3],
    "Num4": [3, 4, 5, "testo", 6, 3, 4, 5, 6, 3, 4, "error", 6, 5, 4],
    "Num5": [10, 20, "venti", 40, 50, 30, 60, 70, "?", 90, 80, 70, 100, 110, 120],
    "Num6": [5, 15, 25, np.nan, 35, 45, "error", 55, 65, 75, 85, "?", 95, "###", 115],
    
    # Variabili categoriche con errori
    "Cat1": ["A", "B", "?", "C", "D", "D", "B", "E", np.nan, "C", "C", "B", "!", "A", "E"],
    "Cat2": ["X", "Y", "Z", "error", "X", "N/A", "Y", "?", "X", "Z", "Z", "X", "###", "Y", "error"],
    "Cat3": ["red", "blue", "green", "error", "?", "yellow", "blue", "missing", "red", "green", "blue", "green", "red", "blue", "error"],

    # Target con errori
    "Target": [1, 0, 1, "N/A", 0, 1, 1, 0, "?", 1, 0, 1, 0, 1, "###"]
}

# Convert to DataFrame
dataset_with_errors = pd.DataFrame(data)
```

```python
df = dataset_with_errors.copy()
```

```python
df
```

```python
df["Num1"] = df["Num1"].replace(["###", "errore"],[np.nan, np.nan])
df["Num2"] = df["Num2"].replace(["?", "N/A"], np.nan)
df["Num3"] = df["Num3"].replace(["missing", "!"], [np.nan, np.nan])
df["Num4"] = df["Num4"].replace(["testo", "error"],[np.nan, np.nan])
df["Num5"] = df["Num5"].replace(["venti", "?"], ["20", np.nan])
df["Num6"] = df["Num6"].replace(["error", "?", "###"], np.nan)
df["Cat1"] = df["Cat1"].replace(["?", "!"], df["Cat1"].mode()[0])
df["Cat2"] = df["Cat2"].replace(["error", "N/A", "###", "?"], df["Cat2"].mode()[0])
df["Cat3"] = df["Cat3"].replace(["?", "error", "missing"], df["Cat3"].mode()[0])

df = df.drop("Target", axis=1)
```

```python
df["target"] = df["Num1"]
```

```python
df = df.drop("Num1", axis=1)
```

```python
from pyparsing import col


df = pd.get_dummies(data=df, columns=["Cat1", "Cat2", "Cat3"], drop_first=True, dtype = int)
```

```python
X_ = df.drop("target", axis=1)
y = df["target"]
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV

from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.compose import ColumnTransformer

from sklearn.pipeline import Pipeline



import joblib
```

```python
X_
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=667)
```

```python
model_lr = LinearRegression().fit(X_train, y_train)
```

```python
model_lr.score(X_train, y_train)
```

```python
model_lr.score(X_test, y_test)
```

```python
cross_val_score(model_lr, X,y, cv=5).mean()
```

```python
model_ridge = Ridge().fit(X_train, y_train)
```

```python
model_ridge.score(X_train, y_train)
```

```python
model_ridge.score(X_test, y_test)
```

```python
print(cross_val_score(model_ridge, X,y, cv=5))
print(cross_val_score(model_ridge, X,y, cv=5).mean())
```

```python
model_rf = RandomForestRegressor().fit(X_train, y_train)
```

```python
print(model_ridge.score(X_train, y_train))

print(model_ridge.score(X_test, y_test))

print(cross_val_score(model_rf, X,y, cv=5))
print(cross_val_score(model_rf, X,y, cv=5).mean())
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators":[100,200,50],
    "min_samples_leaf":[2,5,10],
    "min_samples_split":[2,5,10,15],
    "max_depth":[5,10,15],
    "bootstrap":[True,False],
    "max_features":[4,7]
}

grid = GridSearchCV(
    param_grid=param_grid,
    cv=5,
    estimator=model_rf)


grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
modelforpipe = grid.best_estimator_
```

```python
colonne_categoriche = ["Cat1","Cat2","Cat3"]
```

```python
transformer = ColumnTransformer(transformers = [
    ("encoding", OneHotEncoder(drop = "first", handle_unknown="ignore"), colonne_categoriche)
    ], remainder = "passthrough")

pipe = Pipeline(steps = (
    ("transformer", transformer),
    ("scaler", StandardScaler()),
    ("model", modelforpipe)
))

pipe
```

```python
pipe.fit(X__, y)
```
