```python
df_island = pd.get_dummies(X["island"], dtype=int)
X = pd.concat([X.drop("island", axis=1), df_island], axis=1)
```

```python
X["sex"] = X["sex"].map({"male":1, "female":0})
```

```python
scaler = StandardScaler()
```

```python
X_norm = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
from sklearn.cluster import KMeans

inertia = []
for n in range (1, 11):
    algorithm = (KMeans(n_clusters= n,
                        init= "k-means++",
                        n_init= 10,
                        max_iter= 300,
                        tol= 0.0001,
                        random_state=667,
                        algorithm="elkan"))
    algorithm.fit(X_norm)
    inertia.append(algorithm.inertia_)
```

```python
inertia
```

```python
plt.figure(1 , figsize = (10 ,5))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

### pca

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_norm = pca.fit_transform(X_norm)
```

```python
pca_df = pd.DataFrame(X_norm, columns=[0,1,2])
```

```python
from sklearn.cluster import KMeans

inertia = []
for n in range (1, 11):
    algorithm = (KMeans(n_clusters= n,
                        init= "k-means++",
                        n_init= 10,
                        max_iter= 300,
                        tol= 0.0001,
                        random_state=667,
                        algorithm="elkan"))
    algorithm.fit(pca_df)
    inertia.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (10 ,5))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
algorithm = KMeans(
    n_clusters=3,
    init="k-means++",
    n_init=10,
    max_iter=300,
    tol=0.0001,
    random_state=667,
    algorithm="elkan"
)

algorithm.fit(X_norm)
labels = algorithm.labels_
```

```python
import plotly.graph_objs as go

pca_df["label"] = labels
trace1 = go.Scatter3d(
    x=pca_df[0],
    y=pca_df[1],
    z=pca_df[2],
    mode = "markers",
    marker=dict(
        color = pca_df["label"],
        size= 20,
        line=dict(
            color= pca_df["label"],
            width= 12
    ),
    opacity=0.8)
    )

data = [trace1]
layout = go.Layout(
                    title= 'Clusters',
                    scene = dict(
                            xaxis = dict(title  = '0'),
                            yaxis = dict(title  = '1'),
                            zaxis = dict(title  = '2')
                        )
                )
fig = go.Figure(data=data, layout=layout)
fig.show(render="notebook")
```
