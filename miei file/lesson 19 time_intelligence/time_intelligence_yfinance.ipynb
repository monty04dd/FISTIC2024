```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Scarica i dati di Apple
apple_data = yf.download('AAPL', start='1980-01-01')

# Mantieni la data come indice
# Resample annuale e calcola la media
resampled_data = apple_data.resample('A').mean()

# Aggiungi una colonna per l'anno
resampled_data['Year'] = resampled_data.index.year

# Crea un istogramma con gli anni come colonne
resampled_data.set_index('Year')['Volume'].plot(kind='bar', figsize=(12, 6))
plt.title('Volume Annuale per Anno (Apple)')
plt.xlabel('Anno')
plt.ylabel('Volume medio')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Scarica i dati di Apple
apple_data = yf.download('AAPL', start='1980-01-01')

# Mantieni la data come indice
# Resample annuale e calcola la media
resampled_data = apple_data.resample('YE').mean()

# Aggiungi una colonna per l'anno
resampled_data['Year'] = resampled_data.index.year

# Crea un istogramma con i valori reali
fig, ax = plt.subplots(figsize=(12, 6))
resampled_data.set_index('Year')['Volume'].plot(kind='line', ax=ax)

# Etichette del grafico
ax.set_title('Volume Medio Annuale (Apple)')
ax.set_xlabel('Anno')
ax.set_ylabel('Volume Medio Giornaliero')

# Mostra i valori numerici reali senza notazione scientifica
ax.get_yaxis().get_major_formatter().set_scientific(False)

# Ruota i valori sull'asse X per maggiore leggibilitÃ 
plt.xticks(rotation=45)

plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Scarica i dati di Apple
apple_data = yf.download('AAPL', start='2024-01-01')

# Mantieni la data come indice
# Resample annuale e calcola la media
resampled_data = apple_data.resample('M').mean()

# Aggiungi una colonna per l'anno
resampled_data['Year'] = resampled_data.index.month

# Crea un istogramma con i valori reali
fig, ax = plt.subplots(figsize=(12, 6))
resampled_data.set_index('Year')['Volume'].plot(kind='bar', ax=ax)

# Etichette del grafico
ax.set_title('Volume Medio mensile (Apple)')
ax.set_xlabel('mese')
ax.set_ylabel('Volume Medio Giornaliero')

# Mostra i valori numerici reali senza notazione scientifica
ax.get_yaxis().get_major_formatter().set_scientific(False)

# Ruota i valori sull'asse X per maggiore leggibilitÃ 
plt.xticks(rotation=45)

plt.show()
```
